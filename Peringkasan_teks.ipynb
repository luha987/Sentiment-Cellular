```python
from transformers import pipeline

# Inisialisasi summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Teks yang akan dirangkum
text = """
Artificial intelligence (AI) is a branch of computer science that aims to create machines that can perform tasks that would typically require human intelligence. AI is an interdisciplinary science with multiple approaches, but advancements in machine learning and deep learning are creating a paradigm shift in virtually every sector of the tech industry.
AI has numerous applications, ranging from autonomous vehicles, language translation, image recognition, to drug discovery. It is also raising ethical and security concerns, particularly as its capabilities continue to grow.
"""

# Melakukan summarization
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

# Menampilkan hasil ringkasan
print("Original Text:\n", text)
print("\nSummary:\n", summary[0]['summary_text'])
```

```python
from transformers import pipeline

# Inisialisasi summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Data dummy berbahasa Indonesia
text = """
Indonesia adalah negara kepulauan terbesar di dunia yang terdiri dari lebih dari 17.000 pulau. Negara ini memiliki keanekaragaman budaya, bahasa, dan tradisi yang sangat kaya, dengan lebih dari 300 kelompok etnis dan 700 bahasa daerah.
Indonesia juga terkenal dengan keindahan alamnya, seperti pantai Bali, gunung berapi, dan hutan tropis yang luas. Sebagai negara demokrasi terbesar ketiga di dunia, Indonesia memiliki peran penting dalam geopolitik regional dan internasional.
Namun, negara ini juga menghadapi berbagai tantangan, termasuk perubahan iklim, urbanisasi, dan ketimpangan ekonomi yang cukup besar.
"""

# Melakukan summarization
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

# Menampilkan hasil ringkasan
print("Teks Asli:\n", text)
print("\nRingkasan:\n", summary[0]['summary_text'])
```

QUESTION ANSWERING

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

context = """
Indonesia adalah negara dengan jumlah pulau terbanyak di dunia. Beberapa pulau yang terkenal termasuk Sumatra, Jawa, dan Kalimantan. Ibukota Indonesia adalah Jakarta.
"""

question = "Apa ibukota Indonesia?"
answer = qa_pipeline(question=question, context=context)

print("Pertanyaan:", question)
print("Jawaban:", answer['answer'])
```

PEMODELAN TOPIK

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Data dummy
texts = [
    "Presiden menghadiri pertemuan di Jakarta.",
    "Harga saham naik drastis hari ini.",
    "Banjir besar melanda kota Semarang.",
    "Teknologi AI semakin berkembang pesat."
]

# Preprocessing
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

# Model LDA
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# Menampilkan hasil
for i, topic in enumerate(lda.components_):
    words = [vectorizer.get_feature_names_out()[j] for j in topic.argsort()[-5:]]
    print(f"Topik {i + 1}: {', '.join(words)}")
```

PENGELOMPOKAN DOKUMEN

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

texts = [
    "Presiden Joko Widodo menghadiri pertemuan di Jakarta.",
    "Harga saham naik drastis hari ini.",
    "Banjir besar melanda kota Semarang.",
    "Teknologi AI semakin berkembang pesat."
]

# Preprocessing
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

# KMeans Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

# Menampilkan hasil
for i, label in enumerate(kmeans.labels_):
    print(f"Teks: {texts[i]} => Cluster {label}")
```

TEXT SIMILARITY

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

sentences = [
    "Presiden Joko Widodo menghadiri pertemuan di Jakarta.",
    "Joko Widodo menghadiri sebuah acara di Jakarta."
]

# Encode teks
embeddings = model.encode(sentences)

# Menghitung skor kesamaan
similarity = util.pytorch_cos_sim(embeddings[0], embeddings[1])

print("Teks 1:", sentences[0])
print("Teks 2:", sentences[1])
print("Skor Kesamaan:", similarity.item())
```
